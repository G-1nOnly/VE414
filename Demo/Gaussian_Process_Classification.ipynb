```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.gaussian_process import GaussianProcessClassifier
import matplotlib.pyplot as plt
import numpy as np
X,y=load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
model = GaussianProcessClassifier()

model.fit(X_train[:,:2], y_train)
xx = np.linspace(4, 7, 100)
yy = np.linspace(1, 5, 100).T
xx, yy = np.meshgrid(xx, yy)
Xfull = np.c_[xx.ravel(), yy.ravel()]
probas = model.predict_proba(Xfull)

plt.imshow(probas[:,0].reshape((100, 100)), extent=(4, 7, 1, 5), origin="lower")

plt.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], marker="o", c="r", edgecolor="k")

plt.scatter(X_train[y_train==0, 0], X_train[y_train==0, 1], marker="o", c="b", edgecolor="k")
plt.colorbar()
```
